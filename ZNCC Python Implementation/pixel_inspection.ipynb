```python
from PIL import Image

# Load the image
image_path = '/content/drive/MyDrive/Colab Notebooks/Multiprocessor Programming/im0.png'  # Replace with your image path
image = Image.open(image_path)

# Convert the image to RGB mode (if it's not already in RGB)
image = image.convert('RGBA')

# Get pixel data
pixels = image.load()

# Get the pixel at a specific location (e.g., (x=50, y=100))
x, y = 1, 0
pixel_value = pixels[x, y]
print(f"Pixel value at ({x}, {y}): {pixel_value}")


```

```python
# Optionally, you can print the RGB values of all pixels (be cautious for large images!)
width, height = image.size
for y in range(height):
    for x in range(width):
        pixel_value = pixels[x, y]
        print(f"Pixel value at ({x}, {y}): {pixel_value}")
```
